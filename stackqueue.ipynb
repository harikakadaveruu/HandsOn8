```python
import random
class Stack:
    def __init__(self, size):
        self.stack = [None] * size
        self.top = -1
        self.size = size

    def push(self, item):
        if self.top < self.size - 1:
            self.top += 1
            self.stack[self.top] = item
        else:
            print("Stack Overflow")

    def pop(self):
        if self.is_empty():
            print("Stack Underflow")
        else:
            self.top -= 1
            return self.stack[self.top + 1]

    def is_empty(self):
        return self.top == -1

    def get_top(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            return self.stack[self.top]


class Queue:
    def __init__(self, size):
        self.queue = [None] * size
        self.front = self.rear = -1
        self.size = size

    def enqueue(self, item):
        if (self.is_full()):
            print("Queue Overflow")
        elif (self.front == -1):
            self.front = 0
            self.rear = 0
            self.queue[self.rear] = item
        else:
            self.rear += 1
            self.queue[self.rear] = item

    def dequeue(self):
        if (self.is_empty()):
            print("Queue Underflow")
        elif (self.front == self.rear):
            self.front = -1
            self.rear = -1
        else:
            self.front += 1

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        return self.rear == self.size - 1


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head is None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete_from_beginning(self):
        if self.is_empty():
            print("List is empty!")
            return None
        deleted_item = self.head.data
        self.head = self.head.next
        return deleted_item
#from stack        
stack = Stack(5)
stack.push(1)
stack.push(2)
stack.push(3)
print("Top element in stack:", stack.get_top())
print("Popped element in stack:", stack.pop())
print("Top element after pop in stack:", stack.get_top())
#from queue
queue = Queue(5)
queue.enqueue(1)
queue.enqueue(2)
queue.dequeue()
print("Dequeued element in queue:", queue.dequeue())
#from linkedlist
linked_list = LinkedList()
linked_list.insert_at_beginning(1)
linked_list.insert_at_beginning(2)
linked_list.insert_at_beginning(3)
print("Deleted element in linkedlist:", linked_list.delete_from_beginning())
```
